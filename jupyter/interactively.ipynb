```python
import os
import pandas as pd
```

```python
def load_data(filename):
    data_dir = os.path.join(os.getcwd(), 'data')
    file_path = os.path.join(data_dir, filename)
    if filename.lower().endswith('.csv'):
        return pd.read_csv(file_path)
    elif filename.lower().endswith(('.xlsx', '.xls')):
        return pd.read_excel(file_path)
    else:
        raise ValueError("Unsupported file format")
```

```python
df = load_data('KPI_Report.csv')
```

```python
df['WeekEndingCY'] = pd.to_datetime(df['WeekEndingCY'], format='%m/%d/%Y', errors='coerce')
```

```python
if 'Sales / Day' in df.columns and 'Shop' in df.columns:
    df['Sales / Day'] = pd.to_numeric(df['Sales / Day'].astype(str).str.replace(r'[$,]', '', regex=True), errors='coerce')
    highest_revenue = df.groupby('Shop')['Sales / Day'].mean().sort_values(ascending=False)
    print("Top shops by average daily sales:")
    print(highest_revenue.head())
else:
    print("Required columns not found in data.")
```

```python
if 'BayTime' in df.columns and 'Shop' in df.columns:
    df['BayTime'] = pd.to_numeric(df['BayTime'], errors='coerce')
    lowest_baytime = df.groupby('Shop')['BayTime'].mean().sort_values()
    print("Shops with lowest average BayTime:")
    print(lowest_baytime.head())
```

```python
if 'CPD' in df.columns and 'Shop' in df.columns:
    df['CPD'] = pd.to_numeric(df['CPD'], errors='coerce')
    highest_cpd = df.groupby('Shop')['CPD'].mean().sort_values(ascending=False).round(2)
    print("\nTop shops by average CPD:")
    print(highest_cpd.head())
```

```python
if 'Net Sales - YoY %' in df.columns and 'Shop' in df.columns:
    df['Net Sales - YoY %'] = pd.to_numeric(df['Net Sales - YoY %'].astype(str).str.replace(r'[%\s]', '', regex=True), errors='coerce')
    highest_growth = df.groupby('Shop')['Net Sales - YoY %'].mean().sort_values(ascending=False).round(2)
    print("\nTop shops by average YoY sales growth:")
    print(highest_growth.head())
```

```python
def count_top(df_metric, top_n):
    return df_metric.head(top_n).reset_index()[['Shop']].assign(Appearance=1)

top_1 = pd.concat([
    count_top(highest_revenue, 1),
    count_top(lowest_baytime, 1),
    count_top(highest_cpd, 1),
    count_top(highest_growth, 1)
])

top_3 = pd.concat([
    count_top(highest_revenue, 3),
    count_top(lowest_baytime, 3),
    count_top(highest_cpd, 3),
    count_top(highest_growth, 3)
])

top_5 = pd.concat([
    count_top(highest_revenue, 5),
    count_top(lowest_baytime, 5),
    count_top(highest_cpd, 5),
    count_top(highest_growth, 5)
])

leaderboard = (
    top_1.groupby('Shop').sum().rename(columns={'Appearance': 'Top1'})
    .join(top_3.groupby('Shop').sum().rename(columns={'Appearance': 'Top3'}), how='outer')
    .join(top_5.groupby('Shop').sum().rename(columns={'Appearance': 'Top5'}), how='outer')
    .fillna(0).astype(int)
    .sort_values(by=['Top1', 'Top3', 'Top5'], ascending=[False, False, False])
)

print("\n🏆 Leaderboard:")
print(leaderboard)
```
